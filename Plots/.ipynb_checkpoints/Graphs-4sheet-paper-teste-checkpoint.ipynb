```python
# Initialize variables and DataFrames
#CaseStudy = [1, 2, 3, 4]
CaseStudy=['results_Placement_Natal_Case_1_2_odcs']
yamlFile = ['Placement_Natal_Case_1_2.yaml']
plotpath = [path + "/" + CaseStudy[0],path]

listOdcs=[]

dfCapacities = pd.DataFrame()
dfFiber = pd.DataFrame()
dfOdcs = pd.DataFrame()

noOdcs = 0
listCapacity= 0

doc=[]
for iconfigurations_file in range(len(yamlFile)):
    os.chdir(plotpath[iconfigurations_file])
    with open(yamlFile[iconfigurations_file], 'r') as f:
        doc = np.append(doc,yaml.load(f, Loader=yaml.loader.BaseLoader))
        campaign_name = os.path.splitext(yamlFile[iconfigurations_file])[0] 
        print("Campaign to plot: " + campaign_name)

noSims = int(doc[0]['ShellScriptParameters']['ntasks'])

for isim in range(noSims):
    dfCapacitiesPerSim = pd.DataFrame()
    dfFiberPerSim = pd.DataFrame()
    dfOdcsPerSim = pd.DataFrame()
    for ijobs in range(jobs):
        
        # Read the current capacities
        dfCurrent_capacities = pd.read_csv(plotpath[0]+"/JOB"+str(ijobs)+"/Sim_"+str(isim)+"/df_capacities.csv", usecols=['capacities'])
        dfCurrent_capacities = dfCurrent_capacities.rename(columns={"capacities": "capacities_study"+str(isim)})
        dfCapacitiesPerSim = pd.concat([dfCapacitiesPerSim, dfCurrent_capacities], axis=0)

        # Read the current fiber length
        dfCurrent_fiber = pd.read_csv(plotpath[0]+"/JOB"+str(ijobs)+"/Sim_"+str(isim)+"/df_fiberlength.csv", usecols=['fiberlength'])
        dfCurrent_fiber = dfCurrent_fiber.rename(columns={"fiberlength": "fiberlength_study"+str(isim)})
        dfFiberPerSim = pd.concat([dfFiberPerSim, dfCurrent_fiber], axis=0)

        # Read the current ODCs for Manaus and Natal
        dfCurrent_odcs = pd.read_csv(plotpath[0]+"/JOB"+str(ijobs)+"/Sim_"+str(isim)+"/df_client_association.csv", usecols=['odc_location'])
        dfCurrent_odcs = dfCurrent_odcs.rename(columns={"odc_location": "odc_location_study"+str(isim)})
        dfOdcsPerSim = pd.concat([dfOdcsPerSim, dfCurrent_odcs.drop_duplicates()], axis=0)

         # Calculate the average number of ODCs
        noOdcs = (noOdcs + dfCurrent_capacities.shape[0]) / 2
        
    dfCapacitiesPerSim = dfCapacitiesPerSim.reset_index()
    dfCapacities = pd.concat([dfCapacities,dfCapacitiesPerSim.drop(columns='index')],axis=1)

    dfFiberPerSim = dfFiberPerSim.reset_index()
    dfFiber = pd.concat([dfFiber,dfFiberPerSim.drop(columns='index')],axis=1)

    dfOdcsPerSim = dfOdcsPerSim.reset_index()
    dfOdcs = pd.concat([dfOdcs,dfOdcsPerSim.drop(columns='index')],axis=1)

    listOdcs.append(noOdcs)
```

## Convergence

```python
# Initialize variables and DataFrames
#CaseStudy = [1, 2, 3, 4]
Sims=['5','6','7']
jobs = 1

dfManausConvergenceStudies = pd.DataFrame()
dfNatalConvergenceStudies = pd.DataFrame()

for isim in Sims:
    dfManausConvergence = pd.DataFrame()
    dfNatalConvergence = pd.DataFrame()
    for ijobs in range(jobs):
        # Read the current capacities for Manaus and Natal
        
        dfManausAlgo_convergence = pd.read_csv(plotpathManaus[0]+"/JOB"+str(ijobs)+"/Sim_"+str(isim)+"/dfConvergence.csv", usecols=['generations','avg_obj'])
        dfManausAlgo_convergence = dfManausAlgo_convergence.rename(columns={"generations":"generations_study"+str(isim),"avg_obj": "avg_obj_study"+str(isim)})

        dfNatalAlgo_convergence = pd.read_csv(plotpathNatal[0]+"/JOB"+str(ijobs)+"/Sim_"+str(isim)+"/dfConvergence.csv", usecols=['generations','avg_obj'])
        dfNatalAlgo_convergence = dfNatalAlgo_convergence.rename(columns={"generations":"generations_study"+str(isim),"avg_obj": "avg_obj_study"+str(isim)})
       
        # Concatenate the current DataFrames with the overall DataFrames
        dfManausConvergence = pd.concat([dfManausConvergence, dfManausAlgo_convergence], axis=0)
        dfNatalConvergence = pd.concat([dfNatalConvergence, dfNatalAlgo_convergence], axis=0)
    
    if isim == 5:
        dfManausConvergence = dfManausConvergence.reset_index(drop=True)
        dfManausConvergenceStudies = pd.concat([dfManausConvergenceStudies,dfManausConvergence],axis=0)

        dfNatalConvergence = dfNatalConvergence.reset_index(drop=True)
        dfNatalConvergenceStudies = pd.concat([dfNatalConvergenceStudies,dfNatalConvergence],axis=0)

    else:
        dfManausConvergence = dfManausConvergence.reset_index(drop=True)
        dfManausConvergenceStudies = pd.concat([dfManausConvergenceStudies,dfManausConvergence],axis=1)

        dfNatalConvergence = dfNatalConvergence.reset_index(drop=True)
        dfNatalConvergenceStudies = pd.concat([dfNatalConvergenceStudies,dfNatalConvergence],axis=1)

```

#### Manaus

```python
# Extract unique study identifiers
studies = sorted(set(col.split('_')[2] for col in dfManausConvergenceStudies.columns if 'avg_obj' in col))
label = ['ODCs=RUs/2','ODCs=RUs/3','ODCs=RUs/4']
markers=['o','1','v']

# Plotting
plt.figure(figsize=(10, 6))
i=0
for study in studies:
    generations_col = f'generations_{study}'
    avg_obj_col = f'avg_obj_{study}'
    plt.plot(dfManausConvergenceStudies[generations_col], dfManausConvergenceStudies[avg_obj_col], marker=markers[i], label=label[i])
    i+=1

plt.xlabel('Generations',fontsize='15')
plt.ylabel('Average Objective Value',fontsize='15')
plt.xticks(fontsize='15')
plt.yticks(fontsize='15')
#plt.title('Manaus')
plt.legend(fontsize='15')
plt.grid(True)
plt.savefig("ConvergencePlot_Manaus.png", dpi=150)
plt.show()
```
